```python
import numpy as np
import pandas as pd
```

```python
df1 = pd.DataFrame({'employee': ['Bob', 'Jake', 'Lisa', 'Sue'], 'group': ['Accounting', 'Engineering', 'Engineering', 'HR']})
df1
```

```python
df2 = pd.DataFrame({'name' : ['Jake', 'Bob', 'Lisa', 'Sue', 'Raj'], 'salary' : [70000, 80000, 12000, 90000, 10000]})
df2
```

```python
df3 = pd.merge(df1, df2, left_on='employee', right_on='name')
df3
```

```python
df3.drop('name', axis=1)
```

```python
df = pd.DataFrame({'key' : ['a', 'b', 'c', 'a', 'b', 'c'], 'data1' : range(6)})
df
```

```python
df.groupby('key').sum()
```

```python
df.groupby('key').count()
df.groupby('key').std()
df.groupby('key').min()
df.groupby('key').max()
```

```python
df1 = pd.DataFrame({'key' : ['a', 'b', 'c', 'a', 'b', 'c'], 'data1' : range(6), 'data2' : np.random.randint(0, 10, 6)})
df1
```

```python
df1.groupby('key').aggregate([min, max, np.median])
```

```python
df1.groupby('key').aggregate({'data1' : min, 'data2' : 'max'})
```

```python
df1[df1.data2 > 2]
```

```python
df1.groupby('key').filter(lambda x : x.data2.std() > 3)
```

```python
def banana(x:pd.DataFrame) -> bool:
    return (x.data2.std() > 3)


df1.groupby('key').filter(banana)
```
